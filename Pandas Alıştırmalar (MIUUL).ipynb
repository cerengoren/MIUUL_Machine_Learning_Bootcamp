```python
df.loc[(df["size"]<3) & (df["total_bill"]>10),"total_bill"].mean()
```

```python
df["total_bill_tip_sum"] = df["total_bill"] + df["tip"]
df.head()
```

```python
df.groupby("sex")["total_bill"].mean()
```

```python
f_avg= df[df["sex"]=="Female"]["total_bill"].mean()
f_avg
```

```python
m_avg= df[df["sex"]=="Male"]["total_bill"].mean()
m_avg
```

```python
def func(sex,total_bill):
    if sex =="Female":
        if total_bill < f_avg :
            return 0
        else:
            return 1
        
    else :
        if total_bill < m_avg :
            return 0
        else:
            return 1
```

```python
df["total_bill_flag"] = df.apply(lambda x: func(x["sex"],x["total_bill"]), axis=1)
```

```python
df.head()
```

```python
df.groupby("sex")["total_bill_flag"].value_counts()
```

```python
df.groupby(["sex","total_bill_flag"]).agg({"total_bill_flag":"count"})
```
